```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.pipeline import Pipeline
from custom_imputers import FfillImputer

import warnings
warnings.filterwarnings("ignore")
```

Here's a dataset with dates to sort it by. It has no missing values, so we'll make them

```python
arc = pd.read_csv('data_arc_new.csv', parse_dates=[1, 2])
arc['Активная мощность'][0, 3, 6, 9, 12] = np.nan
arc.head(13)
```

```python
pipe = Pipeline(
    steps=[
        ('imputer', FfillImputer('Активная мощность', 'Начало нагрева дугой'))
    ]
)

trans_df = pipe.fit_transform(arc)
trans_df.head(13)
```

We told it explicitly by which column to order. We could have skipped that, so that the imputer found a datetime column by itself

```python
pipe = Pipeline(
    steps=[
        ('imputer', FfillImputer('Активная мощность'))
    ]
)

trans_df = pipe.fit_transform(arc)
trans_df.head(13)
```

Same result as it actually finds. Pay attention to the first value - it's still missing. That is due to ffill usege by default. We can switch to bfill

```python
pipe = Pipeline(
    steps=[
        ('imputer', FfillImputer('Активная мощность', strategy='bfill'))
    ]
)

trans_df = pipe.fit_transform(arc)
trans_df.head(13)
```

Or alternatively we could stay with ffill, but after that use bfill specially for the first line. I called that 'both_ways'

```python
pipe = Pipeline(
    steps=[
        ('imputer', FfillImputer('Активная мощность', strategy='both_ways'))
    ]
)

trans_df = pipe.fit_transform(arc)
trans_df.head(13)
```
